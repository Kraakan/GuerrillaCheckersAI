```python
import pandas as pd
import json
```

```python
tournament_df = pd.read_excel('../tournament_results.xlsx')
```

```python
tournament_df
```

```python
model_info_file = open("../models/model_info.json", "r")
model_info = json.load(model_info_file)
```

```python
g_indexes = []
c_indexes = []
for key, item in model_info.items():
    if item["player"] == "1":
        g_indexes.append(key)
    if item["player"] == "0":
        c_indexes.append(key)
```

```python
tournament_df["g_index"] = g_indexes
tournament_df = tournament_df.set_index("g_index")
```
